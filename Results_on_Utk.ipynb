```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['CNN1', 'CNN2', 'CNN3', 'CNN4', 'CNN5', 'CNN6', 'CNN7', 'CNN8', 'CNN9', 'CNN10']

# Accuracy values
accuracies = [0.70, 0.45, 0.54, 0.52, 0.51, 0.36, 0.61, 0.65, 0.67, 0.68]

# Precision values
precisions = [0.70, 0.20, 0.55, 0.53, 0.53, 0.38, 0.66, 0.66, 0.68, 0.68]

# Recall values
recalls = [0.70, 0.45, 0.54, 0.52, 0.51, 0.36, 0.61, 0.65, 0.67, 0.68]

# F1-score values
f1_scores = [0.70, 0.28, 0.53, 0.52, 0.51, 0.35, 0.61, 0.65, 0.67, 0.67]

# Convert values to percentages
accuracies = [x * 100 for x in accuracies]
precisions = [x * 100 for x in precisions]
recalls = [x * 100 for x in recalls]
f1_scores = [x * 100 for x in f1_scores]

# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Create the grouped bar chart
plt.figure(figsize=(14, 8))
bars1 = plt.bar(r1, accuracies, color='skyblue', width=bar_width, edgecolor='grey', label='Accuracy')
bars2 = plt.bar(r2, precisions, color='lightgreen', width=bar_width, edgecolor='grey', label='Precision')
bars3 = plt.bar(r3, recalls, color='salmon', width=bar_width, edgecolor='grey', label='Recall')
bars4 = plt.bar(r4, f1_scores, color='gold', width=bar_width, edgecolor='grey', label='F1-Score')

# Add title and labels
plt.title('Model Metrics Comparison on UTK-Face Dataset')
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Scores (%)', fontweight='bold')

# Add xticks on the middle of the group bars
plt.xticks([r + 1.5 * bar_width for r in range(len(models))], models)

# Create legend
plt.legend()

# Add the percentage labels on top of the maximum value bar
def add_max_label(bars, values):
    max_value = max(values)
    for bar in bars:
        height = bar.get_height()
        if height == max_value:
            plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}%', ha='center', va='bottom')

add_max_label(bars1, accuracies)


# Show the graph
plt.show()
```
